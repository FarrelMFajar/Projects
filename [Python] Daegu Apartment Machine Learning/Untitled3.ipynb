```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your level data (adjust the file path as needed)
level_data = pd.read_csv('level_data.csv')

# Convert datetime columns to UTC
level_data['level_start_at'] = pd.to_datetime(level_data['level_start_at'], utc=True)
level_data['level_finish_at'] = pd.to_datetime(level_data['level_finish_at'], utc=True)

# Data Cleaning
# Filter out records where finish time is earlier than start time
level_data_clean = level_data[level_data['level_finish_at'] >= level_data['level_start_at']]

# Handle missing values by filling NaN values with 0 (assuming no power-up usage)
level_data_clean['wild_card_usage'].fillna(0, inplace=True)
level_data_clean['undo_usage'].fillna(0, inplace=True)
level_data_clean['add_card_usage'].fillna(0, inplace=True)

# Ensure that level_result doesn't contain NaN and map 'win' to 1, else 0
level_data_clean = level_data_clean.dropna(subset=['level_result'])
level_data_clean['level_result_numeric'] = level_data_clean['level_result'].apply(lambda x: 1 if x == 'win' else 0)

# Define the power-up usage columns to analyze
usage_columns = ['wild_card_usage', 'undo_usage', 'add_card_usage']

# Function to safely compute correlation
def safe_corr(df, col1, col2):
    if df[col1].isna().all() or df[col2].isna().all():
        return None
    return df[col1].corr(df[col2])

# Correlation and Plotting for each difficulty level
for difficulty in level_data_clean['difficulty_given'].unique():
    plt.figure(figsize=(18, 5))
    
    difficulty_data = level_data_clean[level_data_clean['difficulty_given'] == difficulty]
    
    for i, usage in enumerate(usage_columns, 1):
        plt.subplot(1, 3, i)
        
        # Scatter plot with regression line
        sns.regplot(x=usage, y='level_result_numeric', data=difficulty_data, logistic=True)
        correlation = safe_corr(difficulty_data, usage, 'level_result_numeric')
        
        # Handling potential NaN correlation values
        if correlation is not None:
            plt.title(f"{usage.replace('_', ' ').capitalize()} vs Level Result\n"
                      f"Difficulty: {difficulty} (Correlation: {correlation:.2f})")
        else:
            plt.title(f"{usage.replace('_', ' ').capitalize()} vs Level Result\n"
                      f"Difficulty: {difficulty} (No Correlation)")
        plt.xlabel(f"{usage.replace('_', ' ').capitalize()}")
        plt.ylabel("Level Result (Win=1, Lose=0)")
    
    plt.suptitle(f"Correlation between Power-Up Usage and Level Result for Difficulty {difficulty}", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Correlation and Plotting for all difficulties combined
plt.figure(figsize=(18, 5))

for i, usage in enumerate(usage_columns, 1):
    plt.subplot(1, 3, i)
    
    # Scatter plot with regression line
    sns.regplot(x=usage, y='level_result_numeric', data=level_data_clean, logistic=True)
    correlation = safe_corr(level_data_clean, usage, 'level_result_numeric')
    
    # Handling potential NaN correlation values
    if correlation is not None:
        plt.title(f"{usage.replace('_', ' ').capitalize()} vs Level Result\n"
                  f"All Difficulties (Correlation: {correlation:.2f})")
    else:
        plt.title(f"{usage.replace('_', ' ').capitalize()} vs Level Result\n"
                  f"All Difficulties (No Correlation)")
    plt.xlabel(f"{usage.replace('_', ' ').capitalize()}")
    plt.ylabel("Level Result (Win=1, Lose=0)")

plt.suptitle("Correlation between Power-Up Usage and Level Result for All Difficulties", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
